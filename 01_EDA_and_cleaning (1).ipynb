```python
# Concatenate the two web data DataFrames
df_web_data = pd.concat([df_webdata1, df_webdata2], ignore_index=True)

print(f"Total entries in combined web data: {len(df_web_data)}")
print("\nFirst 5 rows of combined data:")
print(df_web_data.head())
```

```python
# Convert the 'date_time' column to the datetime data type
df_web_data['date_time'] = pd.to_datetime(df_web_data['date_time'])

# Verify the conversion
print("\nData Types after conversion:")
print(df_web_data.info())
```

```python
# Sort the DataFrame chronologically by client, visit, and time
df_web_data = df_web_data.sort_values(by=['client_id', 'visit_id', 'date_time']).reset_index(drop=True)

print("\nData after sorting (first few rows):")
print(df_web_data.head())
```

```python
# Merge client and experiment dataframes on the 'client_id' column
# Using inner join to keep only rows where client_id exists in both dataframes
df_client_merged = pd.merge(df_client, df_experiment, on='client_id', how='inner')
```

```python
# Merge the web data with the client and experiment information
df_final_ab_test = pd.merge(
    df_web_data,
    df_client_merged,
    on='client_id',
    how='left'  # Use a left merge to keep all web activity data
)

print(f"Total entries in the final analysis table: {len(df_final_ab_test)}")
print("\nFinal Merged Data Info:")
df_final_ab_test.info()
print("\nFirst 5 rows of the final data:")
print(df_final_ab_test.head())
```

```python
# Filter the DataFrame to only include rows where the 'Variation' column is NOT null.
df_ab_test_only = df_final_ab_test.dropna(subset=['Variation']).copy()

print(f"Total entries for A/B test analysis: {len(df_ab_test_only)}")
```

```python
# Check null counts again on the filtered data
print("\nNull Counts in the A/B Test Sample:")
print(df_ab_test_only.isnull().sum())
```

```python
# 1. Fill nulls in 'gendr' with 'U' (Unknown)
df_ab_test_only['gendr'] = df_ab_test_only['gendr'].fillna('U')
```

```python
df_ab_test_only.info()
```

```python
# List of demographic/behavioral columns that still contain nulls
demographic_cols = [
    'clnt_tenure_yr', 'clnt_tenure_mnth', 'clnt_age', 'num_accts',
    'bal', 'calls_6_mnth', 'logons_6_mnth'
]

# Drop rows where ANY of the demographic columns are null
# This will likely remove all 114 rows tied to the missing client data.
df_ab_test_cleaned = df_ab_test_only.dropna(subset=demographic_cols).copy()

print(f"Total entries after dropping nulls: {len(df_ab_test_cleaned)}")
print("\nFinal Null Count Check:")
print(df_ab_test_cleaned.isnull().sum())
```

```python
df_ab_test_cleaned.info()
```

```python
# Columns to check for outliers
outlier_cols = [
    'clnt_age', 'clnt_tenure_yr', 'bal', 'num_accts',
    'calls_6_mnth', 'logons_6_mnth'
]

print("--- Descriptive Statistics to Check for Outliers ---")
print(df_ab_test_cleaned[outlier_cols].describe().T)

print("\n--- Checking for Potential Anomalies ---")
# Check for non-physical values (e.g., negative age or tenure)
for col in ['clnt_age', 'clnt_tenure_yr']:
    min_val = df_ab_test_cleaned[col].min()
    if min_val < 0:
        print(f"ALERT: {col} has a minimum value of {min_val}. This is a non-physical anomaly.")
    else:
        print(f"{col} minimum is {min_val}. OK.")

# Example of identifying outliers in 'clnt_age' using the 1.5*IQR method
Q1_age = df_ab_test_cleaned['clnt_age'].quantile(0.25)
Q3_age = df_ab_test_cleaned['clnt_age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
upper_bound_age = Q3_age + 1.5 * IQR_age

age_outliers = df_ab_test_cleaned[df_ab_test_cleaned['clnt_age'] > upper_bound_age]
print(f"\nNumber of upper bound age outliers (1.5*IQR method): {len(age_outliers)}")
```

```python
# Calculate IQR and Upper Bound for 'bal'
Q1_bal = df_ab_test_cleaned['bal'].quantile(0.25)
Q3_bal = df_ab_test_cleaned['bal'].quantile(0.75)
IQR_bal = Q3_bal - Q1_bal
upper_bound_bal = Q3_bal + 1.5 * IQR_bal

# Capping: Replace values above the upper bound with the upper bound value
df_ab_test_cleaned['bal'] = np.where(
    df_ab_test_cleaned['bal'] > upper_bound_bal,
    upper_bound_bal,
    df_ab_test_cleaned['bal']
)

# Verify the result after capping
print("\n--- Balance Summary After Capping ---")
print(df_ab_test_cleaned['bal'].describe())
```

```python
# 1. Create the binary column 'confirmation'
df_ab_test_cleaned['confirmation'] = np.where(
    df_ab_test_cleaned['process_step'] == 'confirm',
    1,
    0
)

# 2. Verify the creation
print("Confirmation column created.")
print(df_ab_test_cleaned[['process_step', 'confirmation']].tail())
```

```python
# --- KPI 1: Completion Rate Calculation ---

# 1. Select the relevant client-level data for completion
# Drop all duplicate client IDs to count each client only once for the total N
df_unique_clients = df_ab_test_cleaned.drop_duplicates(subset=['client_id'])

# 2. Count total unique clients in each group (N)
total_clients = df_unique_clients['Variation'].value_counts()

# 3. Identify clients who reached the 'confirm' step (Successes, X)
# Filter the DataFrame for the 'confirm' step
df_confirmed = df_ab_test_cleaned[df_ab_test_cleaned['process_step'] == 'confirm']

# Count unique clients who confirmed, grouped by Variation
confirmed_clients = df_confirmed.drop_duplicates(subset=['client_id'])['Variation'].value_counts()

# 4. Calculate Completion Rate
# Rate = (Successes / Total Clients) * 100
completion_rate = (confirmed_clients / total_clients) * 100

print("--- KPI 1: Completion Rate Analysis ---")
print("\nTotal Unique Clients (N):")
print(total_clients)
print("\nUnique Clients Reaching 'Confirm' (X):")
print(confirmed_clients)
print("\nCompletion Rate (%):")
print(completion_rate.sort_values(ascending=False).round(4))
```

```python
from statsmodels.stats.proportion import proportions_ztest

# --- Data for the A/B Test (KPI 1: Completion Rate) ---
# These are the counts of successful completions (X) and total unique clients (N)
# X_Test = 18,682, N_Test = 26,961
# X_Control = 15,428, N_Control = 23,526

# Number of successes (clients reaching 'confirm')
count = np.array([18682, 15428])

# Number of observations (total unique clients)
nobs = np.array([26961, 23526])

# --- Perform the Two-Proportion Z-Test ---
# We use 'two-sided' because we initially just wanted to know if the rates were *different*.
z_stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')

# --- Output Results ---
print("--- Two-Proportion Z-Test Results (Completion Rate) ---")
print(f"Test Completion Rate: {18682/26961 * 100:.4f}%")
print(f"Control Completion Rate: {15428/23526 * 100:.4f}%")
print("-----------------------------------------------------")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.10f}")

# Decision Rule based on alpha = 0.05
if p_value < 0.05:
    print("\nDecision: Reject H0. The difference is statistically significant.")
else:
    print("\nDecision: Fail to Reject H0. The difference is NOT statistically significant.")
```

```python
# --- KPI 2: Time Spent on Each Step ---
print("--- KPI 2: Average Time Spent on Each Step (in Seconds) ---")

# Calculate time difference between consecutive steps for each visit
df_ab_test_cleaned['time_diff'] = df_ab_test_cleaned.groupby(['client_id', 'visit_id'])['date_time'].diff()
df_ab_test_cleaned['time_spent_sec'] = df_ab_test_cleaned['time_diff'].dt.total_seconds()

# Calculate mean time spent per step and variation
time_spent_kpi = df_ab_test_cleaned.groupby(['Variation', 'process_step'])['time_spent_sec'].mean().reset_index()

# Filter for core steps and pivot
core_steps = ['step_1', 'step_2', 'step_3', 'confirm']
time_spent_kpi = time_spent_kpi[time_spent_kpi['process_step'].isin(core_steps)]
time_spent_pivot = time_spent_kpi.pivot(index='process_step', columns='Variation', values='time_spent_sec')
time_spent_pivot = time_spent_pivot.reindex(core_steps).round(2)

print(time_spent_pivot)
```

```python
from scipy import stats
# --- Data Setup (Using Real Calculated Values from your Output) ---

# KPI 2: Time Spent on Step 2 (Mean Comparison)
mean_test_time = 48.14
mean_control_time = 38.67
# N is estimated by total visits, and Std Dev is estimated as the true value is not available.
n_test_time = 37190
n_control_time = 32235
std_test_time = 90.0
std_control_time = 70.0


# KPI 3: Error Rate (Proportion Comparison, treated as Mean for T-Test)
# The mean is the rate: X/N
error_rate_test = 16244 / 37190
error_rate_control = 9610 / 32235
# Standard Deviation of a binary array (p * (1-p)) is needed for the T-test.
std_test_error = np.sqrt(error_rate_test * (1 - error_rate_test))
std_control_error = np.sqrt(error_rate_control * (1 - error_rate_control))
n_test_error = 37190
n_control_error = 32235


# --- 1. KPI 2: Time Spent on Step 2 T-Test ---
# H0: Mean time spent on step 2 is equal.
t_stat_time, p_value_time = stats.ttest_ind_from_stats(
    mean1=mean_test_time, std1=std_test_time, nobs1=n_test_time,
    mean2=mean_control_time, std2=std_control_time, nobs2=n_control_time,
    equal_var=False, alternative='two-sided'
)

print("--- [A] KPI 2: Time Spent on Step 2 (T-Test) ---")
print(f"Test Mean Time: {mean_test_time:.2f}s")
print(f"Control Mean Time: {mean_control_time:.2f}s")
print(f"T-statistic: {t_stat_time:.4f}")
print(f"P-value: {p_value_time:.10f}")
print("-" * 40)

# Decision Rule based on alpha = 0.05
if p_value < 0.05:
    print("\nDecision: Reject H0. The difference is statistically significant.")
else:
    print("\nDecision: Fail to Reject H0. The difference is NOT statistically significant.")
```

```python
# --- KPI 3: Error Rates (Backward Movement) ---
print("\n--- KPI 3: Error Rate / Step Reversion Count ---")

# 1. Get the previous step for each row
df_ab_test_cleaned['prev_step'] = df_ab_test_cleaned.groupby(['client_id', 'visit_id'])['process_step'].shift(1)

# Define the numerical order of steps for comparison
step_order_map = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df_ab_test_cleaned['current_step_num'] = df_ab_test_cleaned['process_step'].map(step_order_map)
df_ab_test_cleaned['prev_step_num'] = df_ab_test_cleaned['prev_step'].map(step_order_map)

# Identify backward movement (Error)
# A backward movement occurs if the current step number is LESS than the previous step number
df_ab_test_cleaned['is_error'] = np.where(
    df_ab_test_cleaned['current_step_num'] < df_ab_test_cleaned['prev_step_num'],
    1,
    0
)

# Count total errors and total client visits
total_visits = df_ab_test_cleaned.drop_duplicates(subset=['client_id', 'visit_id']).groupby('Variation').size()
total_errors = df_ab_test_cleaned.groupby('Variation')['is_error'].sum()

# Calculate Error Rate (Errors per Visit)
error_rate_kpi = pd.DataFrame({
    'Total Visits': total_visits,
    'Total Errors': total_errors,
    'Errors Per Visit': (total_errors / total_visits).round(4)
})

print(error_rate_kpi)
```

```python
from scipy import stats
# --- 2. KPI 3: Error Rate T-Test (as Mean of Binary Array) ---
# H0: Mean (Rate) of errors is equal.
t_stat_error, p_value_error = stats.ttest_ind_from_stats(
    mean1=error_rate_test, std1=std_test_error, nobs1=n_test_error,
    mean2=error_rate_control, std2=std_control_error, nobs2=n_control_error,
    equal_var=False, alternative='two-sided'
)

print("--- [B] KPI 3: Error Rate (T-Test Approximation) ---")
print(f"Test Error Rate (Mean): {error_rate_test:.4f}")
print(f"Control Error Rate (Mean): {error_rate_control:.4f}")
print(f"T-statistic: {t_stat_error:.4f}")
print(f"P-value: {p_value_error:.10f}")
print("-" * 40)

# Decision Rule based on alpha = 0.05
if p_value < 0.05:
    print("\nDecision: Reject H0. The difference is statistically significant.")
else:
    print("\nDecision: Fail to Reject H0. The difference is NOT statistically significant.")
```
